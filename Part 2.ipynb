# Part 1

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
filename="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
df=pd.read_html(filename)
```

```python
df=df[0]
df.head()
```

```python
df.shape
```

```python
df.replace("Not assigned",np.nan,inplace=True)
```

```python
df=df[df["Borough"].notna()]
```

```python
df.head()
```

### Here I obtained the data from wikipedia and dropped all rows that had a value of NaN in the Borough column ( after i converted "Not assigned" to NaN.
